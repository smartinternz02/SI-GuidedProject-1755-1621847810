```python
import numpy as np
import pandas as pd
```

# Multi Linear Regression

```python
df  = pd.read_csv("/petrol_consumption.csv")
df.head(5)
```

```python
df.isna().sum()
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=73)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(y_test,y_pred)
print(accuracy)
```

```python
lr.predict([[7.5,5986,1271,0.387]])
```

# Logistic Regression

```python
dataset = pd.read_csv("/bikebuyer1.csv")
```

```python
dataset.isna().sum()
```

```python
dataset["ID"].fillna(dataset["ID"].mean(),inplace=True)
dataset["Children"].fillna(dataset["Children"].mean(),inplace=True)
dataset["Commute Distance"].fillna(dataset["Commute Distance"].mean(),inplace=True)

df = dataset.dropna()

df.isna().sum()
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values

df.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:,1]=le.fit_transform(x[:,1])
x[:,2]=le.fit_transform(x[:,2])
x[:,7]=le.fit_transform(x[:,7])
y=le.fit_transform(y)

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer([("en",OneHotEncoder(),[5,6,10])],remainder="passthrough")
x = ct.fit_transform(x)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
 y_pred = lr.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
print(accuracy)
```
