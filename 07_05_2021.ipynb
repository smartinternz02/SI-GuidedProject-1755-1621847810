```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(sns.get_dataset_names())
```

```python
df = sns.load_dataset("diamonds")
```

```python
df  = df.iloc[:25,:]
```

```python
df.head(10)
```

```python
plt.scatter(df.carat,df.price)
```

```python
df.plot(x="carat",y="price",kind="barh")
```

```python
sns.relplot(data=df,x="carat",y="price",hue="cut")
```

```python
sns.relplot(data=df,x="carat",y="price",hue="cut",col="color",col_wrap=2)
```

```python
sns.distplot(df.price,kde=True)
```

```python
sns.barplot(data=df,x="carat",y="price",hue="cut",ci=100)
```

```python
sns.countplot(data=df,x="color")
```

```python
sns.jointplot(data=df,x="carat",y="price")
```

```python
sns.boxplot(x="price",y="carat",data=df)
```

```python
corr=df.corr()
print(corr)
```

```python
sns.heatmap(corr,annot=True,cmap="PiYG")
```

```python

sns.pairplot(data=df,hue="cut",kind="scatter",palette="husl", markers=["o", "s", "D","8","^"],plot_kws={"s":125})
```
